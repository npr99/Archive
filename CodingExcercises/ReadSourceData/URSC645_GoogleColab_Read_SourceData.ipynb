## Step 4: Read in CSV file using Relative Paths
Relative paths is a powerful tool to help organize a reproducible data science workflow. Relative paths allow for common source files to be shared within a collabartive team project and for code to be shared without needing to change the local directory.

````
import pandas as pd
````

Notice the use of the 

```
'../../'
```

The double dots informs the enviroment to move up one directory relative to the current directory. In the example below the SourceData folder is 2 directories above the current directory.

Once the enviroment is informed about the relative location of the folder the folder path to the data can be set.



````
dataframe = pd.read_csv('../../SourceData/data.census.gov/ACSDT5Y2018.B09002_2020-03-30T094416/ACSDT5Y2018.B09002_data_with_overlays_2020-03-30T094414.csv')
````
